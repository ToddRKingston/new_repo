```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
from pylab import *
```

```python
x = np.linspace(0,5,10)
y=x**2
```

```python
figure()
plot(x,y,'r')
xlabel('x')
ylabel('y')
title('title')
show()
```

```python
subplot(1,2,1)
plot(x,y,'r--')
subplot(1,2,2)
plot(y,x,'g*-');
```

```python
#matplotlib object oriented API
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height (range 0 to 1)

axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

axes2.plot(x,y,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```
